```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
plt.plot([1,3,2,4])
plt.show()   # displaying 
#it generates y or x axis automatically according to needs
```

```python
plt.plot([1,2,3,4],[1,4,9,16])
plt.title('Test Plot', fontsize=8, color='g')
# color of the title can be changed also fontsize
plt.xlabel('number n')
plt.ylabel('squre of n = n^2')
plt.show
```

```python
plt.figure(figsize=(2,5))                   # 2 inches wide x 5 inches tall
plt.plot([1, 2, 3, 4], [1, 4, 9, 16], 'kx') # red-o
plt.axis([0, 6, 0, 20])                     # [xmin, xmax, ymin, ymax]
plt.annotate('square it', (3,6))
plt.show()
```

```python
plt.clf()          # clear figure
x = np.arange(4)   # gives 0,1,2,3 series
y = [8.8, 5.2, 3.6, 5.9]
plt.xticks(x, ('Ankit', 'Hans', 'Joe', 'Flaco'))
# plt.bar(x, y)
# plt.bar(x, y, color='y')
plt.bar(x, y, color=['lime', 'r', 'k', 'tan'])
plt.show()
```

```python
d = {'Red O' : np.random.rand(10),
     'Grn X' : np.random.rand(10)}
df = pd.DataFrame(d)
df.plot(style=['ro','gx'])
plt.show()
```

```python
ts = pd.Series(np.random.randn(180), index=pd.date_range('1/1/2018', periods=180))
df = pd.DataFrame(np.random.randn(180, 3), index=ts.index, columns=list('ABC'))
df.cumsum().plot()
plt.show()
```

```python
N = 50
x = np.random.rand(N)
y = np.random.rand(N)
colors = np.random.rand(N)
sizes = (30 * np.random.rand(N))**2   # 0 to 15 point radii
plt.scatter(x, y, s=sizes, c=colors, alpha=0.5)
plt.show()
```

```python
df = pd.read_csv('Fremont_weather.txt')
print(df)
plt.bar(df['month'], df['record_high'], color='r')
plt.bar(df['month'], df['record_low'], color='c')
plt.plot(df['month'], df['avg_high'], color='k')
plt.plot(df['month'], df['avg_low'], color='b')
plt.legend()
plt.show() 
```

```python
fig = plt.figure()
fig.suptitle('My SubPlots')
fig.add_subplot(221)
plt.plot([np.log(n) for n in range(1,10)])
fig.add_subplot(222, facecolor='y')
fig.add_subplot(223)
fig.add_subplot(224) 
plt.show()
# 221 = top left subplot
# 222 = top right subplot
# 223 = bottom left subplot
# 224 = bottom right subplot
```

```python
fig, plots = plt.subplots(2, sharex=True)
fig.suptitle('Sharing X axis')
x = range(0,200,5)
y = [n**0.8 for n in x]
plots[0].plot(x, y, color='r')
plots[1].scatter(x, y)
plt.show()
```
